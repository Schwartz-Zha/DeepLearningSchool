```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import keras
keras.__version__
```

```python
from keras import models
from keras import layers
from sklearn.model_selection import train_test_split

n_samples=1000
test_size=0.3
random_seed=1
np.random.seed(random_seed)

f=lambda x:np.exp(-x*x)
X=np.random.permutation(np.linspace(-5,5,n_samples))
Y=f(X)

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=test_size)
```

```python
H=4
input_dim=1

model=models.Sequential()

model.add(layers.Dense(H,input_dim=input_dim,kernel_initializer='normal',activation='sigmoid'))

model.add(layers.Dense(1,kernel_initializer='normal',activation='linear'))

model.compile(loss='mean_squared_error',optimizer='adam')
```

```python
model_history=model.fit(X_train,Y_train,batch_size=100,epochs=1000,verbose=1)
```

```python
X_range=np.linspace(-5,5,1000)
y_pred=model.predict(X_range)

fig,ax =plt.subplots(1,1,figsize=(14,8))
ax.scatter(X_train,Y_train,label="Training Data")
ax.plot(X_range,y_pred,lw=4,color='r',label='MLP with one hidden layer')
ax.set_xlabel(r'$X$',fontsize=20)
ax.set_ylabel(r'$Y$',fontsize=20)
ax.set_title('Toy Dataset for Regression',fontsize=24)
ax.tick_params(labelsize=20)

ax.legend(loc=1,fontsize=20)

plt.show()
```
